```python
class File():
    def __init__(self,name,type,size,modified,change,lasta,mode,ino,dev ,nlink,uid,gid,snapshot,dir) -> None:
        super().__init__()
        self.name= name
        self.type=type
        self.size=size
        self.modified = modified# modified time
        self.change = change# when it was change
        self.lasta=lasta# last time access
        self.mode =mode 
        self.ino=ino 
        self.dev = dev
        self.nlink =nlink 
        self.uid = uid 
        self.gid = gid 
        self.snapshot = snapshot
        self.dir = dir
```

```python
def entropy(data):
      e = 0
      counter = collections.Counter(data)
      l = len(data)
      for count in counter.values():
          # count is always > 0
          p_x = count / l
          e += - p_x * math.log2(p_x)
      return e
```
